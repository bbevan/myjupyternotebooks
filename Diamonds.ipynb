# Residuals of a linear model of diamond mass versus Singapore Dollars



```r
library('ggplot2')
```

```r
data(diamonds)

diamond <- sample(diamonds, 10, replace = FALSE)
```

```r
# Assign data to variable names
y <- diamond$price; x <- diamond$carat;

# Generate the model
fit <- lm(y ~ x)
```

```r
# Residuals
e <- resid(fit)
```

```r
# Model
yhat <- predict(fit)
```

```r
# What is the sum of the residuals?
sum(e)
```

```r
plot(diamond$carat, diamond$price,
    xlab = "Mass",
    ylab = "Carats")

abline(fit, lwd=2)
```
